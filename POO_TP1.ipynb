```python
# Computing the correlation between the numerical values of the DataFrame using the plot_correlation_map function
def plot_correlation_map(df): 
 # Select the numerical columns only 
   numeric_columns = df.select_dtypes(include=['number']) 
   # Calcul of the correlation matrix for the numerical columns
   corr = numeric_columns.corr()
   # Specify the figure's size
   plt.figure(figsize=(12, 10)) 
   # Choose the color palette 
   cmap = sns.diverging_palette(220, 10, as_cmap=True) 
   # Trace a HeatMap that allows us to visualise the correlation between different values of the DtaFrame
   sns.heatmap(corr, cmap=cmap, square=True, cbar_kws={'shrink': 0.9}, annot=True, annot_kws={'fontsize': 12}) 
# Display the HeatMap 
plot_correlation_map(df)
```

```python
# The 'plot_correlation_map()' function aims to create a visual representation of the correlation between variables,in order to evaluate how they are related to each other and it can help identify which pairs of variables tend to move together
# The intensity of the color indicates the strength of the correlation
# A good positif correllation equals a strong correlation coefficient (close to 1), and means that two values increase in a linear manner ; as one variable increases, the other tends to increase as well
# A good negative correlation equals a correlation coefficient that's close to -1, which reflects a linear but opposite relationship where as one variable increases, the other tends to decrease
# According to the heatmap and the previous statements, the best correlation are the ones between the columns :
# Parch and Fare
# Parch and SibSp
# Parch and Age
# Pclass and Fare
# Pclass and Age
```

```python
# Creating a new column named 'Title' to assign the appropriate title for each passenger using the column 'Name'

import pandas as pd
df=pd.read_csv(R'C:/Users/hp/OneDrive/Bureau/tp/titanic-passengers.csv', sep=';')

# Define the Title Dictionary

Title_Dictionary = {
    "Capt": "Officer",
    "Col": "Officer",
    "Major": "Officer",
    "Dr": "Officer",
    "Rev": "Officer",
    "Jonkheer": "Royalty",
    "Don": "Royalty",
    "Sir": "Royalty",
    "Lady": "Royalty",
    "the Countess": "Royalty",
    "Dona": "Royalty",
    "Mme": "Miss",
    "Mlle": "Miss",
    "Miss": "Miss",
    "Ms": "Mrs",
    "Mr": "Mrs",
    "Mrs": "Mrs",
    "Master": "Master"
}

# Use the function 'apply' to apply a function to each element of the column
# The 'lambda' function extracts the title from the name in the expected format and uses the `Title_Dictionary' dictionary to assign the appropriate title
# If the title is not found in the dictionary, "Unknown" is set as a default value

df['Title'] = df['Name'].apply(lambda x: Title_Dictionary.get(x.split(",")[1].split(".")[0].strip(), "Unknown"))

# Display the DataFrame with the new column "Title"
df
```
